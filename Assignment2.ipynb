-  What is a hash table, and how does it relate to dictionaries in  Python?
- A hash table is a data structure that stores key–value pairs for fast lookups.
-  A dictionary is Python’s implementation of a hash table.

-  When you store data in a dictionary:

-  The key is passed through a hash function.

-  The hash function produces a hash value.

-  That value maps to an index (bucket) in the hash table.

-  The value is stored in that bucket, associated with the key.

- How does the “in” keyword work differently for lists and dictionaries?
- the in keyword behaves a bit differently depending on the data structure.
-** For Lists**
- The in keyword checks if a value exists inside the list.
- It goes through elements one by one (linear search, O(n)).



```python
fruits = ["apple", "banana", "mango"]

print("apple" in fruits)
print("grape" in fruits)
```

- For Dictionaries
- The in keyword only checks keys by default, not values.
- Lookup is very fast (O(1) on average, because dictionaries use hash tables).

```python
student = {"name": "nehra", "age": 21, "city": "Delhi"}

print("name" in student)
print("Alice" in student)
```

- What is a nested dictionary, and give an example of its use case?
- A nested dictionary in Python is simply a dictionary inside another dictionary.
- The outer dictionary contains keys that map to inner dictionaries.
- This allows you to represent hierarchical or structured data.
exaple:-
- students = {
    "101": {"name": "Alice", "age": 21, "course": "Math"},
    "102": {"name": "Bob", "age": 22, "course": "Physics"},
    "103": {"name": "Charlie", "age": 20, "course": "Computer Science"}
}

print(students["102"]["name"])    Output: Bob
print(students["103"]["course"])  Output: Computer Science

- Why are dictionaries considered unordered, and how does that affect data retrieval?
- Traditionally, a dictionary in Python is implemented as a hash table.
Keys are stored based on their hash values, not their insertion sequence.
- Retrieval is by key, not position
- You don’t do dict[0] like a list.
- You must know the key.
- student = {"name": "sangeeta", "age": 21, "city": "Delhi"}
- print(student["age"])    works
- print(student[0])      Error
- Duplicates are disallowed
- If you insert the same key again, the previous value gets replaced, not reordered.
- d = {"a": 1, "b": 2, "a": 3}
- print(d)  # {'a': 3, 'b': 2}
